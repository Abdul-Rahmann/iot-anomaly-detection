#### Summary Statistics

```python
import glob

normal_files = glob.glob("../data/processed/fan/id_00/normal/*.parquet")
anomaly_files = glob.glob("../data/processed/fan/id_00/abnormal/*.parquet")

df_normal = pd.concat([pd.read_parquet(f) for f in normal_files])
df_anomaly = pd.concat([pd.read_parquet(f) for f in anomaly_files])

print(f"Total Normal Samples: {df_normal.shape[0]}")
print(f"Total Anomalous Samples: {df_normal.shape[0]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

num_features = 13
n_rows = 100

num_cols = 4
num_rows = math.ceil(num_features / num_cols)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 5, num_rows * 4))
axes = axes.flatten()

for i in range(num_features):
    sns.kdeplot(df_normal.iloc[:n_rows, i], label="Normal", fill=True, ax=axes[i])
    sns.kdeplot(df_anomaly.iloc[:n_rows, i], label="Anomaly", fill=True, color="red", ax=axes[i])
    axes[i].set_title(f"{df_anomaly.columns[i]}")
    axes[i].set_xlabel(f"Feature {i + 1}")
    axes[i].set_ylabel("Density")
    axes[i].legend()

for j in range(num_features, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

#### Feature Selection

```python
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(df_anomaly)

# Compute cumulative explained variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker="o")
plt.axhline(y=0.90, color="r", linestyle="--", label="90% Variance")
plt.axhline(y=0.95, color="g", linestyle="--", label="95% Variance")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA: Choosing the Optimal Number of Components")
plt.legend()
plt.grid()
plt.show()
```

```python
# Retain 95% of variance
variance_threshold = 0.95

# Find the number of components needed to retain variance_threshold%
n_components = np.argmax(cumulative_variance >= variance_threshold) + 1
print(f"Number of PCA Components Retaining {variance_threshold*100}% Variance: {n_components}")

pca = PCA(n_components=n_components)
df_normal_pca = pca.fit_transform(df_normal)
df_anomaly_pca = pca.transform(df_anomaly)
```
